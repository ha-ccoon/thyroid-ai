```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
data1 = pd.read_csv('./data/allhypo.data')
data1 = pd.DataFrame(data1)
data1
```

```python
# 컬럼 이름 설정
columns = ['age', 'sex', 'on thyroxine', 'query on thyroxine', 'on antithyroid medication', 'sick', 'pregnant', 'thyroid surgery', 'I131 treatment', 'query hypothyroid', 'query hyperthyroid', 'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych', 'TSH measured', 'TSH', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U measured', 'T4U',  'FTI measured', 'FTI', 'TBG measured', 'TBG', 'referral source', 'misc']
data1.columns = columns
data1
```

```python
# 컬럼 정리
data1[['diagnosis', 'id']] = data1['misc'].str.split(r'\.\|', expand=True)

data1
```

```python
data1.drop(columns=['misc'], inplace=True)
data1
```

```python
# 결측값 컬럼 삭제
data1.drop(columns=['TBG'], inplace=True)
data1
```

```python
# 사용하지 않는 컬럼 삭제
data1.drop(columns=['T3 measured', 'TT4 measured', 'T4U measured', 'FTI measured', 'TBG measured', 'TSH measured', 'referral source'], inplace=True)
data1
```

```python
# diagnosis 분리
# diagnsis = data1['diagnosis']
# id = data1['id']
```

```python
# 결측값 정리
data1.replace('?', 0, inplace=True)
data1.fillna(0, inplace=True)
data1
```

```python
data1.replace({'f': 0, 't': 1}, inplace=True)
data1
```

```python
# sex
data1['sex'] = data1['sex'].replace({'M': 0, 'F': 1})

data1
```
